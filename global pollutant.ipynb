```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('global_air_pollution_dataset.csv')
df.columns = df.columns.str.strip()  # Clean any extra spaces

# Use only pollutants available in your dataset
pollutants = ['PM2.5 AQI Value', 'NO2 AQI Value', 'CO AQI Value', 'Ozone AQI Value']

# If you have a 'Year' column, otherwise use another column for x-axis
years = df['Year'] if 'Year' in df.columns else df.index

for pollutant in pollutants:
    if pollutant in df.columns:
        plt.plot(years, df[pollutant], marker='o', label=pollutant)

plt.title("Yearly Average Pollutant Trends (2021-2023)")
plt.xlabel("Year" if 'Year' in df.columns else "Index")
plt.ylabel("Average Concentration")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('global_air_pollution_dataset.csv')
df.columns = df.columns.str.strip()

# Use only columns that exist in your dataset!
columns = ['PM2.5 AQI Value', 'NO2 AQI Value', 'CO AQI Value', 'Ozone AQI Value']

for col in columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

fig, axs = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(df[columns[0]], kde=True, bins=30, color='red', ax=axs[0, 0])
axs[0, 0].set_title('Distribution of PM2.5')
axs[0, 0].set_xlabel('PM2.5')
axs[0, 0].set_ylabel('Count')

sns.histplot(df[columns[1]], kde=True, bins=30, color='blue', ax=axs[0, 1])
axs[0, 1].set_title('Distribution of NO2')
axs[0, 1].set_xlabel('NO2')
axs[0, 1].set_ylabel('Count')

sns.histplot(df[columns[2]], kde=True, bins=30, color='green', ax=axs[1, 0])
axs[1, 0].set_title('Distribution of CO')
axs[1, 0].set_xlabel('CO')
axs[1, 0].set_ylabel('Count')

sns.histplot(df[columns[3]], kde=True, bins=30, color='orange', ax=axs[1, 1])
axs[1, 1].set_title('Distribution of Ozone')
axs[1, 1].set_xlabel('Ozone')
axs[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('global_air_pollution_dataset.csv')
df.columns = df.columns.str.strip()

# Use columns available in your dataset
columns = ['PM2.5 AQI Value', 'NO2 AQI Value']  # Change as needed

def get_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

for i, col in enumerate(columns):
    df[col] = pd.to_numeric(df[col], errors='coerce')
    outliers = get_outliers(df[col])
    axs[i].scatter(df.index, df[col], c='blue', label='Data')
    axs[i].scatter(df.index[outliers], df[col][outliers], c='red', label='Outliers')
    axs[i].set_title(f'{col} Outliers')
    axs[i].set_xlabel('Index')
    axs[i].set_ylabel(col)
    axs[i].legend()

plt.tight_layout()
plt.show()

```
